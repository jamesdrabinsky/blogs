# Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv('../Labs/2.01-lab-pandas-titanic/train-Copy1.csv')
```

I will be demonstrating the following pandas cleaning practices:
- Changing the index
- Dropping columns 
- Rename columns
- Use df.map to change all values in a column

```python
#First drop all nulls

df.dropna(inplace=True)
```

### Chaging the Index

```python
#Set PassngerID as index
df.set_index('PassengerId', inplace=True)
```

```python
df.head()
```

### Dropping Columns

```python
# Drop Parch and Tickets Columns
df.drop(columns=['Parch', 'Ticket'], inplace=True)
```

```python
df.head()
```

### Rename Column

```python
# Rename Pclass as just Class
df.rename(columns=({'Pclass': 'Class'}), inplace=True)
```

```python
df.head()
```

### Use df.map

```python
# Convert all floats in Age column to intergers
df['Age'] = df['Age'].map(lambda x: int(x))
```
