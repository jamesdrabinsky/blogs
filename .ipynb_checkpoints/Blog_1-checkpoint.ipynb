# Pandas Concatenation

```python
import pandas as pd


import warnings
warnings.filterwarnings('ignore')
```

```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
one = pd.DataFrame(
    {'student_id':['1','2','3','4','5'],
     'Marks_scored':[98,90,87,69,78]},
    index=['James', 'John', 'Joe', 'Jerry', 'Jacob'])

two = pd.DataFrame(
    {'student_id':['6','7','8','9','10'],
     'age':[5,7,8,7,9]},
    index=['Billy', 'Bill', 'Berry', 'Bryce', 'Betty'])

display_side_by_side(one, two)
```

### Pandas Concatenation Parameters

**objs** - list the objects that you want to concatenate.

```python
#Ex.
pd.concat([one,two])
```

**axis** - specify whether you want to concatenate along rows (0) or columns (1).

```python
#Ex.
pd.concat([one,two], axis=1) #0 is the default
```

**join** - you can set this paramater to:

- inner: this will only retain rows in which both tables share a key value.
- left: keeps all the values that are in the left (primary) table
- right: keeps all the values that are in the right table.
- outer: retais all rows occuring in both tables.

```python
#Ex.
pd.concat([one,two], axis=0, join='inner')
```

**ignore_index**: if you want to keep the original row labels.

```python
#Ex.
pd.concat([one,two], axis=0, join='outer', ignore_index=True)
```
